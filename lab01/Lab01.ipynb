```python
import numpy as np
import matplotlib.pyplot as plt
from collections import OrderedDict
```

## Q1

```python
def gener1(a, b, m, seed):
    arr = []
    x = seed
    arr.append(x)
    while True:
        x = (a * x + b) % m
        if(x == seed):
            break
        arr.append(x)
    return arr
```

```python
print("sequence for a=6, b=0, m=11, and,")
for i in range(0, 11):
    print("seed =", i, end=": ")
    print(gener1(6, 0, 11, i))
```

```python
print("sequence for a=3, b=0, m=11, and,")
for i in range(0,11):
    print("seed =", i, end=": ")
    print(gener1(3, 0, 11, i))
```

## Q2

```python
def gener2(a, b, m, seed):
    x = seed
    u = x/m
    ui = []
    xi = []
    # ui.append(u)
    # xi.append(x)
    mapp = dict()
    checker = dict()
    for i in range(0, 20):
        mapp[i] = 0
    for _ in range(10000):
        x = (a*x + b) % m
        u = x/m
        if x in checker:
            break
        else:
            checker[x]=1
        xi.append(x)
        ui.append(u)

        key = round(u*20) % 20
        mapp[key] += 1
        
    return  mapp

ranges = ['0, 0.05',
          '0.05, 0.1',
          '0.1, 0.15',
          '0.15, 0.2',
          '0.2, 0.25',
          '0.25, 0.3',
          '0.3, 0.35',
          '0.35, 0.4',
          '0.4, 0.45',
          '0.45, 0.5',
          '0.5, 0.55',
          '0.55, 0.6',
          '0.6, 0.65',
          '0.65, 0.7',
          '0.7, 0.75',
          '0.75, 0.8',
          '0.8, 0.85',
          '0.85, 0.9',
          '0.9, 0.95',
          '0.95, 1']
```

#### a = 1597, b = 1, m = 244944, seed = 1

```python
mapp1 = gener2(1597, 1, 244944, 1)
mapp1 = OrderedDict(sorted(mapp1.items()))

for i in range(0, 20):
    print(ranges[i], ":", mapp1[i])
```

```python
keys = list(mapp1.keys())

newMapp1 = dict()
for (key, value) in mapp1.items():
    newMapp1[round(key*0.05, 2)] = value

x = list(newMapp1.keys())
y = list(newMapp1.values())

plt.figure(figsize=(18, 5))
width = np.min(np.diff(keys))/3

plt.bar(keys - 0.5*width, y, color='green', width = 0.3, label='a = 1597, seed = 1')
plt.xticks(keys, ranges, rotation='vertical')
plt.title('a = 1597, b = 1, m = 244944, seed = 1')
plt.show()
```

## Q3

```python
def gener3(a, b, m, seed):
    x = seed
    u = x/m
    ui = []
    ui.append(u)
    checker = dict()
    checker[x]=1
    for _ in range(10000):
        x = (a*x + b) % m
        u = x/m
        if (x==0): continue
        if x in checker:
            break
        else:
            checker[x]=1
        ui.append(u)
        
    return ui
```

```python
u1 = gener3(1229, 1, 2048, 1)
u2 = gener3(1229, 1, 2048, 1)

del u1[len(u1)-1]
del u2[0]
```
